```python
%matplotlib inline
#1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(10,10))   #(width,height)

#3. plot the data
ax.plot(x,y)

#4. customize plot
ax.set(title='simple plot',xlabel='x-axis',ylabel='y-axis')

#5. save and show(you save the whole figure)
fig.savefig('sample_plot.png')
```

### making figures with numpy arrays

```python
x = np.linspace(0,10,num=100)
x
```

```python
#line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# bar plot
prices = {'almond_butter':10,
         'peanut_butter':8,
         'cashew_butter':12}
fig, ax = plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title='shri\'s nut butter store',ylabel='price($)',
       xlabel='[butter_type]');
# plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()),list(prices.values()));
```

```python
# histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### two options for subplots

```python
# method-1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(prices.keys(),prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# method-2 (recommended)
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(prices.keys(),prices.values());
ax[1,1].hist(np.random.randn(1000));
```
