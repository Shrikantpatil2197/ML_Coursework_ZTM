```python
# bar plot
prices = {'almond_butter':10,
         'peanut_butter':8,
         'cashew_butter':12}
fig, ax = plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title='shri\'s nut butter store',ylabel='price($)',
       xlabel='[butter_type]');
# plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()),list(prices.values()));
```

```python
# histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### two options for subplots

```python
# method-1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(prices.keys(),prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# method-2 (recommended)
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(prices.keys(),prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
#pyplot method
over50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6));
```

```python
# object oriented method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           ax=ax);
ax.set_xlim([45,100])
```

### object oriented method from scratch

```python
over50.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over50['age'],
                    y=over50['chol'],
                    c=over50['target']);

ax.set(title='heart-disease data',
      xlabel='age',
      ylabel='cholesterol');

ax.legend(*scatter.legend_elements(),title='Target');

ax.axhline(over50['chol'].mean(),linestyle='--')
```

```python
over50.target.values
```

```python
over50.head()
```

```python
fig, (ax0,ax1) = plt.subplots(figsize=(10,10),nrows=2,ncols=1);
scatter = ax0.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target']);
ax0.set(title='cholesterol levels',
       xlabel='age',
       ylabel='cholesterol');
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over50['chol'].mean(),
           linestyle='--');

scatter2 = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target']);
ax1.set(title='max-heart levels',
       xlabel='age',
       ylabel='max-heart rate');
ax1.legend(*scatter2.legend_elements(),title='Target');
ax1.axhline(over50['thalach'].mean(),
           linestyle='--');
```

```python
#how to share x-axis(age) and title to whole figure
fig, (ax0,ax1) = plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True);
scatter = ax0.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target']);
ax0.set(title='cholesterol levels',
       ylabel='cholesterol');
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over50['chol'].mean(),
           linestyle='--');

scatter2 = ax1.scatter(x=over50['age'],
                      y=over50['thalach'],
                      c=over50['target']);
ax1.set(title='max-heart levels',
       xlabel='age',
       ylabel='max-heart rate');
ax1.legend(*scatter2.legend_elements(),title='Target');
ax1.axhline(over50['thalach'].mean(),
           linestyle='--');

fig.suptitle('Heart-Disease Analysis',fontsize=16,fontweight='bold');
fig.savefig('Heart-Disease Analysis')
```
